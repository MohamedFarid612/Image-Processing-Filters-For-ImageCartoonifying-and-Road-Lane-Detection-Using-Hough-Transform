```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

img = cv2.imread("C:\\Users\\Dell\\Desktop\\gg.png")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img)
plt.show()  

median = cv2.medianBlur(img, ksize=5)
plt.imshow(median)
plt.show()

gaussian = cv2.GaussianBlur(median,ksize=(5,5),sigmaX=0,sigmaY=0)
plt.imshow(gaussian)
plt.show()


gray = cv2.cvtColor(gaussian, cv2.COLOR_RGB2GRAY)
plt.imshow(gray, cmap='gray')
plt.show()

laplacian = cv2.Laplacian(gray, cv2.CV_64F , ksize=3)
plt.imshow(laplacian, cmap='gray')
plt.show()


# make white img of size img.shape
white = np.ones(img.shape, dtype=np.uint8) * 255
# make black pixels in black of laplacian
white[laplacian < 10] = 0
plt.imshow(white)
plt.show()

# scale_factor = 10
# laplacian = scale_factor * laplacian
# laplacian = cv2.convertScaleAbs(laplacian) 
# ret, thresh = cv2.threshold(laplacian, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
# plt.imshow(thresh, cmap='gray')
# plt.show()





thresh = 255 - white
plt.imshow(thresh, cmap='gray')
plt.show()

size = 5 
repeat = 10  
color_strength = 10 
positional_strength =10 
plt.imshow(img)
plt.show()  
for i in range(repeat):
    img = cv2.bilateralFilter(img, size, color_strength, positional_strength)

plt.imshow(img)
plt.show()



black = np.zeros(img.shape, dtype=np.uint8)
plt.imshow(black)
plt.show()

black[thresh == 255] = img[thresh == 255]
plt.imshow(black)
plt.show()
```
